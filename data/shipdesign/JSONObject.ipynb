# Generate JSON OBject from Flat File

This file reads a MS Excel Table and converts it into a JSON objects.

```python
#Import Libraries
import pandas as pd
import numpy as np 
import json as jsn
```

```python
#Read file and set edge as index
df1 = (pd.read_excel('SampleDBnodes.xlsx')).set_index('objects')
df1
```

```python
#Format to JSON

#Add columnheaders as keys and format:
columnlist = list(df1.columns.values)
for column in columnlist:
    df1[column] = '"'+column+'":"' + df1[column].astype(str)+'"'

df1['keys'] = df1[columnlist].apply(lambda row: ','.join(row.values.astype(str)), axis=1)
df1['keys'] = '{'+df1['keys'].astype(str)+'}'

#Combine any redundant edges and close with hard brackets
df1 = (df1.groupby(['objects'])['keys'].agg(lambda x: ','.join(x.dropna()))).reset_index()
df1['keys'] = '['+df1['keys'].astype(str)+']'

#Format edge and combine with node to create a single column that will converted to a string
df1['objects'] = '"'+df1['objects'].astype(str)+'"'
df1['entry'] = df1.apply(lambda row: ':'.join(row.values.astype(str)), axis=1)
stringjson = '{'+df1.entry.str.cat(sep=',')+'}'
```

```python
#Create json object from string
json_object = jsn.dumps(jsn.loads(stringjson), indent=4)
print(json_object)

# Writing to sample.json
#with open("edges.json", "w") as outfile:
with open("nodes.json", "w") as outfile:
    outfile.write(json_object)
```
